```python
import numpy as np
import matplotlib.pyplot as plt
# Generate simulated temperature readings from two sensors.
# Sensor A: mean 25 C, std 3 C, 200 readings.
# Sensor B: mean 27 C, std 4.5 C, 200 readings.
# Also generate 200 timestamps uniformly from 0 to 10 seconds.
# Use np.random.default_rng with a seed = last 4 digits of your Drexel ID.

seed = 1234  # Replace with last 4 digits of your Drexel ID
rng = np.random.default_rng(seed)

# Generate temperature readings
temps_a = rng.normal(loc=25.0, scale=3.0, size=200)
temps_b = rng.normal(loc=27.0, scale=4.5, size=200)

# Generate timestamps
timestamps = rng.uniform(low=0.0, high=10.0, size=200)
```

# Scatter plot: sensor reading (y-axis) vs timestamp in seconds (x-axis).
# Color Sensor A points blue and Sensor B points orange.
# Add axis labels with units, a title, and a legend.

```python
plt.figure(figsize=(10, 6))

# Plot sensor A and B as scatter plots
plt.scatter(timestamps, sensor_a, color='blue', label='Sensor A', alpha=0.6)
plt.scatter(timestamps, sensor_b, color='orange', label='Sensor B', alpha=0.6)

# Add labels and title
plt.xlabel('Time (seconds)')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Readings from Two Sensors')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Overlaid histogram of Sensor A and Sensor B temperature distributions.
# Use 30 bins, alpha=0.5 for transparency so both distributions are visible.
# Add vertical dashed lines at each sensor's mean.
# Include a legend labeling each sensor.
```

```python
plt.figure(figsize=(10, 6))

# Overlaid histograms with transparency
plt.hist(sensor_a, bins=30, alpha=0.5, color='blue', label='Sensor A')
plt.hist(sensor_b, bins=30, alpha=0.5, color='orange', label='Sensor B')

# Add vertical dashed lines at each sensor's mean
mean_a = sensor_a.mean()
mean_b = sensor_b.mean()
plt.axvline(mean_a, color='blue', linestyle='--', linewidth=2, label=f'Mean A: {mean_a:.2f}°C')
plt.axvline(mean_b, color='orange', linestyle='--', linewidth=2, label=f'Mean B: {mean_b:.2f}°C')

# Labels and title
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.title('Temperature Distribution: Sensor A vs Sensor B')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Side-by-side box plot comparing Sensor A and Sensor B distributions.
# Label x-axis with sensor names, y-axis with "Temperature (deg C)".
# Add a horizontal dashed line at the overall mean of both sensors combined.
```

```python
plt.figure(figsize=(10, 6))

# Create side-by-side box plots
plt.boxplot([sensor_a, sensor_b], labels=['Sensor A', 'Sensor B'], patch_artist=True)

# Calculate overall mean of both sensors
overall_mean = np.mean(np.concatenate([sensor_a, sensor_b]))

# Add horizontal dashed line at overall mean
plt.axhline(overall_mean, color='red', linestyle='--', linewidth=2, label=f'Overall Mean: {overall_mean:.2f}°C')

# Labels and title
plt.ylabel('Temperature (°C)')
plt.title('Temperature Distribution Comparison: Box Plot')
plt.legend()
plt.grid(True, alpha=0.3, axis='y')
plt.show()
```
